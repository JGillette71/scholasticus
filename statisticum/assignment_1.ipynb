### Problem 5

What are the advantages and disadvantages of a very flexible (versus a less flexible) approach for regression or classification? Under what circumstances might a more flexible approach be preferred to a less flexible approach? When might a less flexible approach be preferred?

The advantage of a more flexible model is that it often results in a more accurate fit in both regression and classification problems. In the case of a regression problem, we may employ a more flexible model that better fits all samples in training data while in classification, the more flexible method may fit a more accurate decision boundary. However, there is a trade-off between flexibility and interpretability, leading to less explainable relationships between variables for inference problems. What's more is that highly flexible models are prone to over-fitting training data. Thus we may seek a less flexible model in order to better generalize predictions on unseen test data. 

### Problem 6

Describe the differences between a parametric and a non-parametric statistical learning approach. What are the advantages of a parametric approach to regression or classification (as opposed to a nonparametric approach)? What are its disadvantages?

Parametric methods refer to a class of methods where the problem of estimating systemic information $f(x)$ is reduced to estimating a set of parameters applied to input variables. This is often seen as making an informed assumption about the underlying functional form, e.g., linear, logistic, etc., then applying a weight $\\b_p$ to each of the inputs variable to reflect its relative influence on the response variable. Parametric methods contrast non-parametric methods because there is no assumption of the underlying form, thus non-parametric method are significantly more flexible, but also more complex. The advantage of parametric methods is the increased simplicity and interpretability. However, parametric methods may be inadequate for more complex data as assumed functional form may be a poor fit.
